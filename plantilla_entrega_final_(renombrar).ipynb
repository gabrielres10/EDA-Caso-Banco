Según el diccionario de datos, estos -1 significan ausencia del objeto, es decir, no cuenta con carros o hijos. Siendo así, se remplazará por el valor 0.

```python
class NonNegativeIntVariablesTransformer(BaseEstimator, TransformerMixin):
  """Convierte los valores negativos en 0 de los valores enteros que no
  pueden ser negativos por el contexto del problema"""
  def __init__(self):
    self.non_negative_int_variables = non_negative_int_variables

  def fit(self, X, y=None):
    return self

  def transform(self, X, y=None):
    X_transformed = X.copy()
    X_transformed[self.non_negative_int_variables] = X[self.non_negative_int_variables].replace(-1, 0).astype(int)
    return X_transformed
```

## Ingreso bruto menor a ingreso líquido
Debido a las reglas del negocio, que el ingreso bruto sea menor al ingreso líquido es una inconsistencia. Por esto debemos buscar la manera de corregirlo. En el momento en el que se presenta esta inconsistencia, asumiremos que uno de los dos ingresos es real y el otro es el incorrecto. Para decidir cuál es cuál, seguiremos la siguiente regla:

```python
X_transformed_train["log_ingreso_liquido"] = np.log(X_transformed_train["ingreso_liquido"])
X_transformed_train.loc[X_transformed_train["log_ingreso_liquido"] == -np.Infinity, "log_ingreso_liquido"] = None
sns.set(style="darkgrid")

# creating a figure composed of two matplotlib.Axes objects (ax_box and ax_hist)
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})

# assigning a graph to each ax
sns.boxplot(X_transformed_train["log_ingreso_liquido"], ax=ax_box)
sns.histplot(data=X_transformed_train, x="log_ingreso_liquido", ax=ax_hist)

# Remove x axis name for the boxplot
ax_box.set(xlabel='')
plt.show()
```

1. Sacamos el logaritmo del ingreso líquido para remover el sesgo a la izquierda que tiene la variable.
2. Los valores que tenían 0 en ingreso líquido quedan como -infinito, los pondremos como null para no incluirlos en el estudio.
3. Si el logaritmo del ingreso líquido está por encima del rango intercuartil, consideraremos que el ingreso bruto es el valor real y el ingreso líquido será el valor erróneo. En caso contrario, consideraremos el ingreso líquido como el valor real.

Para definir el límite del rango intercuartil, ignoraremos los clientes que tienen el ingreso líquido erróneo.

```python
q3, q1 = np.percentile(X_transformed_train.loc[X_transformed_train["log_ingreso_liquido"] < 18, "log_ingreso_liquido"], [75, 25])
X_transformed_train.drop(columns=["log_ingreso_liquido"], inplace=True)
iqr = q3 - q1
log_limite_iqr = q3 + 1.5*iqr
print(log_limite_iqr)
```

```python
limite_iqr = np.exp(log_limite_iqr)
print(limite_iqr)
```

Teniendo en cuenta esto, el límite de ingreso líquido para considerarlo real es de 12.825.915

Pondremos toda esta lógica en un Transformer para el pipeline de procesamiento.

```python
class NullDirtyIncomeDataTransformer(BaseEstimator, TransformerMixin):

  def fit(self, X, y=None):
    X["log_ingreso_liquido"] = np.log(X["ingreso_liquido"])
    X.loc[X["log_ingreso_liquido"] == -np.Infinity, "log_ingreso_liquido"] = None
    q3, q1 = np.percentile(X.loc[X["log_ingreso_liquido"] < 18, "log_ingreso_liquido"], [75, 25])
    X.drop(columns=["log_ingreso_liquido"], inplace=True)
    iqr = q3 - q1
    log_limite_iqr = q3 + 1.5*iqr
    self.limite_iqr = np.exp(log_limite_iqr)
    return self

  def transform(self, X, y=None):
    # Perform arbitary transformation
    X_transformed = X.copy()
    X_transformed.loc[np.logical_and(X_transformed["ingreso_bruto"] < X_transformed["ingreso_liquido"], X_transformed["ingreso_liquido"] > self.limite_iqr), "ingreso_liquido"] = None
    X_transformed.loc[X_transformed["ingreso_bruto"] < X_transformed["ingreso_liquido"], "ingreso_bruto"] = None
    return X_transformed
```

Entrenaremos dos regresiones lineales, una para imputar el valor de ingreso bruto cuando contemos con el ingreso líquido, y otra para imputar el valor del ingreso líquido cuando contemos con el ingreso bruto. Ya que se van a entrenar con los mismos datos, ambas regresiones serán la función inversa de la otra.

```python
ingreso_bruto_model = LinearRegression()
ingreso_liquido_model = LinearRegression()
ingreso_liquido = X_transformed_train.loc[X_transformed_train.notna().all(axis=1), "ingreso_liquido"]
ingreso_bruto = X_transformed_train.loc[X_transformed_train.notna().all(axis=1), "ingreso_bruto"]
ingreso_bruto_model.fit(np.array(ingreso_liquido).reshape(-1, 1), ingreso_bruto)
ingreso_liquido_model.fit(np.array(ingreso_bruto).reshape(-1, 1), ingreso_liquido)
ingreso_bruto_r_sq = ingreso_bruto_model.score(np.array(ingreso_liquido).reshape(-1, 1), ingreso_bruto)
ingreso_liquido_r_sq = ingreso_liquido_model.score(np.array(ingreso_bruto).reshape(-1, 1), ingreso_liquido)
m, b = ingreso_bruto_model.coef_[0], ingreso_bruto_model.intercept_
print(f"R^2 Ingreso Bruto -> Ingreso Liquido: {ingreso_liquido_r_sq}")
print(f"R^2 Ingreso Liquido -> Ingreso Bruto: {ingreso_bruto_r_sq}")
plt.plot(ingreso_liquido, ingreso_bruto, 'o')
plt.plot(ingreso_liquido, m*ingreso_liquido+b)
plt.show()
```

Vemos cómo se ajusta la regresión lineal a los datos, contamos con un R^2 de 0.86. Ahora procederemos a crear un Transformer que impute los datos faltantes de ingresos.

```python
class CleanIncomeDataTransformer(BaseEstimator, TransformerMixin):

  def fit(self, X, y=None):
    self.ingreso_bruto_model = LinearRegression()
    self.ingreso_liquido_model = LinearRegression()
    ingreso_liquido = X.loc[X.notna().all(axis=1), "ingreso_liquido"]
    ingreso_bruto = X.loc[X.notna().all(axis=1), "ingreso_bruto"]
    self.ingreso_bruto_model.fit(np.array(ingreso_liquido).reshape(-1, 1), ingreso_bruto)
    self.ingreso_liquido_model.fit(np.array(ingreso_bruto).reshape(-1, 1), ingreso_liquido)
    return self

  def transform(self, X, y=None):
    # Perform arbitary transformation
    X_transformed = X.copy()
    if X_transformed["ingreso_bruto"].isna().sum() > 0:
      X_transformed.loc[X_transformed["ingreso_bruto"].isna(), "ingreso_bruto"] = self.ingreso_bruto_model.predict(np.array(X_transformed.loc[X_transformed["ingreso_bruto"].isna(), "ingreso_liquido"]).reshape(-1, 1))
    if X_transformed["ingreso_liquido"].isna().sum() > 0:
      X_transformed.loc[X_transformed["ingreso_liquido"].isna(), "ingreso_liquido"] = self.ingreso_liquido_model.predict(np.array(X_transformed.loc[X_transformed["ingreso_liquido"].isna(), "ingreso_bruto"]).reshape(-1, 1))
    return X_transformed
```

```python
class DummyTransformer(BaseEstimator, TransformerMixin):
  def __init__(self):
    self.categorical_cols = categorical_cols
    categories = tuple(list(translation_dict[col].values()) for col in self.categorical_cols)
    self.encoder = OneHotEncoder(categories=categories)

  def fit(self, X, y=None):
    self.encoder.fit(X[self.categorical_cols])
    return self

  def transform(self, X, y=None):
    # Perform arbitary transformation
    X_transformed = X.copy()
    encoded_data = encoder.transform(X_transformed[self.categorical_cols]).toarray()
    columns = encoder.get_feature_names_out(self.categorical_cols)
    dummies = pd.DataFrame(encoded_data, columns=columns, index=X_transformed.index).astype(int)
    dummies = dummies.drop(columns=[col+"_sin_informacion" for col in self.categorical_cols])
    X_transformed = pd.concat([X_transformed, dummies], axis = 1)
    X_transformed.drop(columns=self.categorical_cols, inplace=True)
    X_transformed.tiene_propiedades = X_transformed.tiene_propiedades.astype(int)
    return X_transformed
```

Creamos un objeto Transformer que haga el clustering por nosotros y que dummifique la variable de clustering, puesto que esta será una simple categoría del cluster al que pertenece.

```python
class ClusteringTransformer(BaseEstimator, TransformerMixin):
  def __init__(self, n_clusters, covariance_type):
    self.n_clusters = n_clusters
    self.covariance_type = covariance_type

  def fit(self, X, y=None):
    self.clustering_columns = [x for x in X.columns if x.startswith('PC')]
    self.clustering_columns.append('LDA')
    self.gmm = GaussianMixture(self.n_clusters, covariance_type=self.covariance_type, random_state=0)
    self.gmm.fit(X[self.clustering_columns])
    return self

  def transform(self, X, y=None):
    # Perform arbitary transformation
    X_transformed = X.copy()
    dummie_clustering = pd.get_dummies(pd.Series(self.gmm.predict(X_transformed[self.clustering_columns]), index=X_transformed.index))
    for i in range(self.n_clusters):
      if i not in dummie_clustering.columns:
        dummie_clustering[str(i)] = 0
    dummie_clustering = dummie_clustering.add_prefix('GMM')
    dummie_clustering = dummie_clustering.drop([f'GMM{self.n_clusters-1}'],axis=1)
    X_transformed = pd.concat([X_transformed,dummie_clustering],axis=1)
    return X_transformed
```

# Data Pipeline

Creamos un método que nos faciliten la creación de pipeline de procesamiento de datos con los distintos parámetros que tienen los pasos.

```python
def create_pipeline(num_pca, num_clusters, covariance_type, lasso_c):
  num_pca = int(round(num_pca))
  num_clusters = int(round(num_clusters))
  type_covariance = "spherical"
  if covariance_type < 0.33:
    type_covariance = "tied"
  elif covariance_type < 0.66:
    type_covariance = "full"
  data_pipeline = [
      ('CategoricalVariablesTranslation', CategoricalVariablesTranslationTransformer()),
      ('NonNegativeIntVariables', NonNegativeIntVariablesTransformer()),
      ('NullDirtyIncomeData', NullDirtyIncomeDataTransformer()),
      ('Scaler', ScalerTransformer()),
      ('CleanIncomeData', CleanIncomeDataTransformer()),
      ('Dummy', DummyTransformer()),
      ('PCA', PCATransformer(num_pca)),
      ('LDA', LDATransformer()),
      ('Clustering', ClusteringTransformer(num_clusters, type_covariance)),
      ('LassoVariableSelection', LassoVariableSelectionTransformer(lasso_c))
  ]

  data_pipeline = Pipeline(steps=data_pipeline)
  return data_pipeline
```

Ahora creamos un método que nos añada el modelo que escojamos al pipeline

```python
def add_model(data_pipeline, model) -> Pipeline:
    whole_pipeline = Pipeline([
        ("data_pipeline", data_pipeline),
        ("model", model)
    ])
    return whole_pipeline
```

# Modelamiento

Iniciaremos con la creación de modelos de predidcción. Para esto, realizaremos los siguientes pasos:
1. Haremos cross validation sobre el set de entrenamiento haciendo uso de optimización bayesiana, esto con la intención de buscar cuál modelo funciona mejor con pocos datos.
2. Del resultado del paso anterior, escogeremos los modelos que mejor se comporten y haremos nuevamente una búsqueda de hiperparámetros con optimización bayesiana, esta vez sin cross validation (entrenando con el train set y probando con el validation set).
3.  Del resultado del paso anterior, escogeremos el mejor modelo junto con sus hiperparámetros y lo evaluaremos con el set de test. Esto nos dará información sobre qué tan bien generaliza el modelo los datos.
4. Con este modelo, lo reentrenaremos con todos los datos disponibles para nos desperdiciar datos.
5. Se generarán las predicciones para realizar el submission.

La función que se optimizará será la siguiente:
$$
  f(val\_auc, train\_auc) = (-ln(1-val\_auc)) ^ {val\_auc/train\_auc}
$$
Aquí vemos que al optimizar la funcion f, haremos que el val_auc crezca al mismo tiempo que la proporción entre el val_auc y el train_auc sea mayor. Es decir, estamos minimizando el overfitting y el underfitting.

```python
def cross_validate_pipeline(pipeline):
  kf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)

  roc_auc_train = []
  roc_auc_val = []

  # K-Fold cross val
  for i, (train_index, test_index) in enumerate(kf.split(X_train, y_train)):
      kX_train, kX_val = X_train.iloc[train_index], X_train.iloc[test_index]
      ky_train, ky_val = y_train.iloc[train_index], y_train.iloc[test_index]
      pipeline.fit(kX_train, ky_train.values.ravel())

      val_preds = pipeline.predict_proba(kX_val)[:, 1]
      val_auc = roc_auc_score(ky_val, val_preds)

      train_preds = pipeline.predict_proba(kX_train)[:, 1]
      train_auc = roc_auc_score(ky_train, train_preds)

      roc_auc_train.append(train_auc)
      roc_auc_val.append(val_auc)

  roc_auc_val = np.array(roc_auc_val)
  roc_auc_train = np.array(roc_auc_train)

  adj_val_roc_auc = ( roc_auc_val.mean() -  roc_auc_val.std())
  adj_train_roc_auc = ( roc_auc_train.mean() -  roc_auc_train.std())
  print(f"Validation ROC AUC adjusted score: {adj_val_roc_auc}")
  print(f"Train ROC AUC adjusted score: {adj_train_roc_auc}")
  f = lambda x, y: ( -np.log(1-x) ) ** ( x / y )
  return pipeline, f(adj_val_roc_auc, adj_train_roc_auc)
  # return pipeline, adj_val_roc_auc
```

```python
def optimize(target_func, pbounds, n_iter):
  pbounds_data_pipeline = {'num_pca': (2, 13), 'num_clusters': (2, 7), 'covariance_type': (0, 1), 'lasso_c': (0.01, 1)}
  pbounds = {**pbounds, **pbounds_data_pipeline}
  np.random.seed(42)
  optimizer = BayesianOptimization(
      f=target_func,
      pbounds=pbounds,
      random_state=1,
      verbose=2
  )
  optimizer.maximize(
    init_points=5,
    n_iter=n_iter,
  )
  return optimizer
```

## **Redes Neuronales - 0.9498**

Ahora entraremos en el mundo de las redes neuronales, aquí contaremos con muchos hiperparámetros, como lo son: el número de capas ocultas y su número de neuronas, el learning_rate inicial, el alpha que penaliza el overfitting, el tamaño del batch y el número máximo de iteraciones.

Para la arquitectura de la red, se definió que todas las capas tendrán una potencia de 2 como número de neuronas, donde la última capa oculta tendrá 8 neuronas y de ahí hacia atrás, cada capa tendrá el doble de neuronas que la siguiente. Por ejemplo, si valor de hidden_layer_size_exp es de 8, tendremos 8 - 3 = 5 capas ocultas, con la siguiente arquitectura: (128, 64, 32, 16, 8).

```python
def train_and_evaluate_mlp_classifier(num_pca, num_clusters, covariance_type, lasso_c,
                                      hidden_layer_size_exp, lr_init, alpha, batch_size, max_iter):
    data_pipeline = create_pipeline(num_pca, num_clusters, covariance_type, lasso_c)
    max_exponent = int(hidden_layer_size_exp)
    model_kwargs = dict(
        hidden_layer_sizes = [2**n for n in reversed(range(3, max_exponent))],
        solver="adam",
        batch_size=2**int(batch_size),
        learning_rate_init=10**lr_init,
        alpha=10**alpha,
        max_iter=int(max_iter),
        early_stopping=True,
        random_state=42,
    )
    model = MLPClassifier(**model_kwargs)

    pipeline = add_model(data_pipeline, model)
    return cross_validate_pipeline(pipeline)

def target_func(**kwargs):
    model, result = train_and_evaluate_mlp_classifier(**kwargs)
    return result

pbounds = {'hidden_layer_size_exp': (4, 7), 'lr_init': (-2, 0), 'alpha': (-3, 0), 'batch_size': (6, 12), 'max_iter': (200, 600)}
optimizer = optimize(target_func, pbounds, n_iter=50)
```
